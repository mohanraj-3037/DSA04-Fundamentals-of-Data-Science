```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('bodyfat.csv')
print(df.head())
```

```python
print("The mean  Age and \n Bodyfat is:",df.mean())
print("The median Bodyfat is:",df.median())
print("The Standard Deviation Bodyfat is:",df.std())
```

```python
plt.boxplot([df['age'], df['fat(%)']], labels=['Age', 'Body Fat %'])
plt.title("Distribution of Age and Body Fat")
plt.show()
```

```python
plt.scatter(df['age'], df['fat(%)'],marker="o")
plt.title("Age vs Body Fat")
plt.xlabel("Age")
plt.ylabel("Body Fat %")
plt.show()
```
