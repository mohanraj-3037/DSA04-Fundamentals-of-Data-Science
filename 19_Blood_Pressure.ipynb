```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df=pd.read_csv('bp.csv')
print(df.head())
```

```python
df['BP_Difference']=df['Systolic_BP']-df['Diastolic_BP']
print(df.head())
```

```python
print("Mean Blood Pressure is:",df['BP_Difference'].mean())
```

```python
group=df.groupby('Treatment_Group')
print(group['BP_Difference'].value_counts())
```

```python
dev=df['BP_Difference'].std()
print(dev)
```

```python
n=df['BP_Difference'].count()
dr=df[df['Treatment_Group']=="Drug A"].count()
pb=df[df['Treatment_Group']=="Placebo"].count()
print(n)
print(dr)
print(pb)
```

Standard Error

```python
SE=dev/np.sqrt(n)
print(SE)
```

```python
t_critical=stats.t.ppf(q=0.975,df=n-1)
print(t_critical)
```

Margin of Error

```python
Mar=t_critical*SE
print(Mar)
```

Confidence Interval


```python
Mean=df['BP_Difference'].mean()
Interval=(Mean-Mar,Mean+Mar)
print(Interval)
```

Confidence Interval of Drug A

```python
drug = df[df['Treatment_Group'] == 'Drug A']

mean_d = drug['BP_Difference'].mean()
std_d  = drug['BP_Difference'].std(ddof=1)
n_d    = drug['BP_Difference'].count()

SE_d = std_d / np.sqrt(n_d)
ME_d = 1.96 * SE_d

CI_drug = (mean_d - ME_d, mean_d + ME_d)
print("Drug A CI:", CI_drug)
```

Confidence Interval of Placebo

```python
drug = df[df['Treatment_Group'] == 'Placebo']

mean_d = drug['BP_Difference'].mean()
std_d  = drug['BP_Difference'].std(ddof=1)
n_d    = drug['BP_Difference'].count()

SE_d = std_d / np.sqrt(n_d)
ME_d = 1.96 * SE_d

CI_drug = (mean_d - ME_d, mean_d + ME_d)
print("Drug A CI:", CI_drug)
```
