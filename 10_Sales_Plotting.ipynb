```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sales.csv')
print(df.head())
```

```python
plt.bar(df["Month_sales"], df["Price"])
plt.xlabel("Month_sales")
plt.ylabel("Price")
plt.title("Monthly Sales Bar Plot")
plt.show()
```

```python
monthly = df.groupby("Month_sales")["Price"].mean()

plt.plot(monthly.index, monthly.values)
plt.xlabel("Month_sales")
plt.ylabel("Price")
plt.title("Monthly Sales Line Plot")
plt.show()
```
