```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('grocerystore.csv')
print(df.head())
```

```python
total=df['Sales'].sum()
print("The total Customer Sales is:",total)
```

```python
discount = (df['Sales'] *( df['Discount']/100))
a_d=df['Sales']-discount
print(a_d)
```

```python
tax=a_d*(df['Tax']/100)
print(tax)
```

```python
Final=a_d+tax
print("The Final Amount after deduced with Discount and Tax is:",Final)
```
