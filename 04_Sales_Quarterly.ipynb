```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('sales.csv')
print(df.head())
```

```python
year = df[df['Month_sales'].str.contains("2023")]
total_sales_2023 = year['Total'].sum()
print("The total sales held in the year 2023 is:",total_sales_2023)
```

```python
df['Month_sales_num'] = df['Month_sales'].str.split(' ').str[0].map({
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
})

Q1 = df[df['Month_sales_num'].isin([1, 2, 3])]['Total'].sum()
Q4 = df[df['Month_sales_num'].isin([10, 11, 12])]['Total'].sum()

total_sales_year = df['Total'].sum()

if Q1 != 0:
    percentage_increase = ((Q4 - Q1) / Q1) * 100
else:
    percentage_increase = 0 # or handle as appropriate if Q1 can be zero

print("Total sales for the year:", round(total_sales_year, 2))
print("Percentage increase from Q1 to Q4:", round(percentage_increase, 2), "%")
```
