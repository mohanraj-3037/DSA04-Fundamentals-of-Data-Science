```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('order.csv')
print(df.head())
```

```python
orders_per_customer = df.groupby('Id').size()
print(orders_per_customer)
```

```python
orders_per_product = df.groupby('Items').size()
print(orders_per_product)
```

```python
df['Order'] = pd.to_datetime(df['Order'])

earliest_date = df['Order'].min()
latest_date = df['Order'].max()

print("Earliest Order Date:", earliest_date)
print("Latest Order Date:", latest_date)
```
