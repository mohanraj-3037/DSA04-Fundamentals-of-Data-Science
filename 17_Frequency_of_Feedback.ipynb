```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('creview.csv')
print(df.head())
```

```python
review=df['summary'].str.lower().str.split().explode().value_counts()
remove=["the","and","is"]
rem_w= review[~review.index.isin(remove)]
print(rem_w)
```

```python
print(rem_w.nlargest(5))
```

```python
review=rem_w.nlargest(5)
plt.bar(review.index,review.values)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Most Frequent Words")
plt.show()
```
